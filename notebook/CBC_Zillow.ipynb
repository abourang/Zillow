```python
import pandas as pd

# Define the paths to the CSV files
file_path_2016 = '/Users/Charlotte/Desktop/Data analytics project/Raw data/zillow-prize-1/properties_2016.csv'
file_path_2017 = '/Users/Charlotte/Desktop/Data analytics project/Raw data/zillow-prize-1/properties_2017.csv'

# Load the CSV files into DataFrames
df_2016 = pd.read_csv(file_path_2016)
df_2017 = pd.read_csv(file_path_2017)

# Concatenate the DataFrames
combined_df = pd.concat([df_2016, df_2017], ignore_index=True)

# Display the first few rows of the combined DataFrame to verify it loaded correctly
print("First few rows of the combined DataFrame:")
print(combined_df.head())
```

```python

def characterize_dataframe(combined_df):
    # Get basic info
    print("Basic Information:")
    print(combined_df.info())
    print("\n")
    
    # Generate descriptive statistics
    print("Descriptive Statistics:")
    print(combined_df.describe(include='all'))
    print("\n")
    
    # Check for missing values
    print("Missing Values:")
    print(combined_df.isnull().sum())
    print("\n")
    
    # Identify unique values and data types
    print("Unique Values and Data Types:")
    for column in combined_df.columns:
        print(f"Column: {column}")
        print(f"Data Type: {combined_df[column].dtype}")
        print(f"Number of Unique Values: {combined_df[column].nunique()}")
        print(f"Unique Values: {combined_df[column].unique()}")
        print("\n")

# Example usage
# df = pd.read_csv('your_file.csv')
characterize_dataframe(combined_df)
```

```python

def missing_values_percentage(combined_df):
    # Calculate the percentage of missing values for each column
    missing_percentage = combined_df.isnull().mean() * 100
    
    # Create a DataFrame for better readability
    missing_percentage_df = pd.DataFrame(missing_percentage, columns=['Missing Percentage'])
    
    # Print the DataFrame
    print("Percentage of Missing Values Per Column:")
    print(missing_percentage_df)
    
# Example usage
# df = pd.read_csv('your_file.csv')
missing_values_percentage(combined_df)
```

```python

def remove_high_missing_columns(df, threshold=98):
    # Calculate the percentage of missing values for each column
    missing_percentage = df.isnull().mean() * 100
    
    # Identify columns to drop based on the threshold
    columns_to_drop = missing_percentage[missing_percentage > threshold].index
    
    # Drop the columns
    df_cleaned = df.drop(columns=columns_to_drop)
    
    return df_cleaned

def save_cleaned_dataframe(df, file_name):
    # Save the cleaned DataFrame to a CSV file
    df.to_csv(file_name, index=False)
    print(f"DataFrame saved to {file_name}")

# Example usage
# df = pd.read_csv('your_file.csv')
df_cleaned = remove_high_missing_columns(combined_df)
save_cleaned_dataframe(df_cleaned, 'cleaned_data.csv')
```

```python


# Remove duplicate rows
df_no_duplicates = combined_df.drop_duplicates()

# Save the cleaned DataFrame
df_no_duplicates.to_csv('data_no_duplicates.csv', index=False)
print("Duplicates removed and DataFrame saved as 'data_no_duplicates.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_no_duplicates is your DataFrame without duplicates

# Visual inspection using box plots for numerical columns
numerical_columns = df_no_duplicates.select_dtypes(include=['int64', 'float64']).columns

# Plot box plots for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df_no_duplicates[column])
    plt.title(f'Box plot for {column}')
    plt.show()

# Visual inspection using scatter plots for pairs of numerical columns (example for first two pairs)
for i in range(min(2, len(numerical_columns))):  # Adjust the range to inspect more pairs if needed
    for j in range(i+1, min(2, len(numerical_columns))):
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=df_no_duplicates[numerical_columns[i]], y=df_no_duplicates[numerical_columns[j]])
        plt.title(f'Scatter plot for {numerical_columns[i]} vs {numerical_columns[j]}')
        plt.xlabel(numerical_columns[i])
        plt.ylabel(numerical_columns[j])
        plt.show()
```
